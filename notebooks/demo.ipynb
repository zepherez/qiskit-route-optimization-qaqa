# Quantum-Assisted Mobility Optimization (Ruter-inspired)

This notebook demonstrates a simplified mobility routing problem inspired by public transportation systems such as Ruter.

We model a small transport network in Oslo and solve a weighted routing optimization problem using QAOA (Quantum Approximate Optimization Algorithm).


```python
import numpy as np

from src.data import travel_time, demand, stops
from src.mobility_model import build_mobility_qp
from src.qaoa_solver import solve_qaoa
```

## Problem Context

We model a simplified public transport routing problem:

- Nodes = transport stops in Oslo
- Edges = travel time between stops
- Weights = passenger demand per stop

The goal is to find a route that minimizes travel time while prioritizing high-demand stops.



```python
print("Stops:")
for s in stops:
    print("-", s)

print("\nTravel time matrix (minutes):")
print(travel_time)

print("\nPassenger demand per stop:")
print(demand)
```

## Step 1: Optimization Model

We formulate the problem as a Quadratic Program (QP), where:

- Binary variables represent stop order
- Constraints ensure each stop is visited exactly once
- Objective minimizes weighted travel time


```python
qp = build_mobility_qp(travel_time, demand)

print("Quadratic program created:")
print(qp.prettyprint())
```

## Step 2: Quantum Optimization (QAOA)

We solve the optimization problem using QAOA via Qiskit.

This is a hybrid quantum-classical algorithm that approximates solutions for combinatorial optimization problems.


```python
result = solve_qaoa(qp)

print("Optimization result:")
print(result)
```

```python
def decode_solution(result, n, stops):
    x = result.x
    route = {}

    idx = 0
    for i in range(n):
        for t in range(n):
            if x[idx] > 0.5:
                route[t] = i
            idx += 1

    return [stops[route[t]] for t in sorted(route.keys())]


route = decode_solution(result, len(stops), stops)

print("\nOptimized mobility route:")
for i, stop in enumerate(route):
    print(f"{i+1}. {stop}")
```

## Result Interpretation

The output represents an optimized transport route between selected Oslo stops.

This is a simplified prototype of:

- Route optimization in public transport networks
- Demand-aware mobility planning
- Hybrid quantum-classical optimization workflows


## Limitations

- Small-scale problem (3 stops only)
- Static travel times (no real-time traffic updates)
- No full quantum advantage demonstrated at this scale

This project is intended as a research prototype for exploring quantum-assisted mobility optimization concepts.
